```python
import math
from math import sqrt
from qiskit import*
```

------------TUGAS NOMOR 4 (A) ------------

```python
qc = QuantumCircuit(3)
qc.h([0,2])
qc.x(1)
qc.barrier()
qc.cx(0,1)
qc.z(2)
qc.z(0)
qc.barrier()
qc.z([0,2])
qc.cx(0,1)
qc.barrier()
qc.h([0,2])
qc.x(1)
display(qc.draw(output = "mpl" ))
```

-------------TUGAS NOMOR 4 (B) -------------

```python
qc = QuantumCircuit(3)
qc.h([0,1,2])
qc.barrier()
qc.x(1)
qc.h(2)
qc.ccx(0,1,2)
qc.x(1)
qc.h(2)
qc.barrier()
display(qc.draw(output = "mpl" ))
```

-------------TUGAS NOMOR 4 (C) -------------

```python
qc = QuantumCircuit(4)
qc.h([0,3])
qc.x(1)
qc.barrier()
qc.cx(0,1)
qc.z(2)
qc.cz(1,3)
qc.barrier()
qc.ccx(3,2,0)
qc.swap(1,3)
qc.cx(2,3)
qc.cswap(3,1,0)
display(qc.draw(output = "mpl" ))
```
